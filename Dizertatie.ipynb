```python
import matplotlib.pyplot as plt
import matplotlib.animation as ani
import numpy as np
import time
start = time.time()
zile = 0
  

GREY = (0.78, 0.78, 0.78)  # neinfectati
RED = (0.96, 0.15, 0.15)  # infectati
GREEN = (0, 0.86, 0.03)  # vindecati
BLACK = (0, 0, 0)  # decedati

COVID19_PARAMS = {
    "r0": 2.28,
    "incubare": 5,
    "stare_medie": 0.8,
    "vindecare": (7, 14),
    "sever": 0.2,
    "vindecare_severa": (21, 42),
    "moarte_severa": (14, 56),
    "rata_mortalitate": 0.034,
    "serial_interval": 7
}


class Virus():
    def __init__(self, params):
        # creare spatiu de animare
        self.fig = plt.figure()
        self.axes = self.fig.add_subplot(111, projection="polar")
        self.axes.grid(False)
        self.axes.set_xticklabels([])
        self.axes.set_yticklabels([])
        self.axes.set_ylim(0, 1)

        # creare etichete
        self.ziua = self.axes.annotate(
            "Ziua 0", xy=[np.pi / 2, 1], ha="center", va="bottom")
        self.infectati = self.axes.annotate(
            "Infectati: 0", xy=[3 * np.pi / 2, 1], ha="center", va="top", color=RED)
        self.decedati = self.axes.annotate(
            "\nDecedati: 0", xy=[3 * np.pi / 2, 1], ha="center", va="top", color=BLACK)
        self.vindecati = self.axes.annotate(
            "\n\nVindecati: 0", xy=[3 * np.pi / 2, 1], ha="center", va="top", color=GREEN)

        # declarare variabile de baza
        self.day = 0
        self.total_nr_infectati = 0
        self.nr_infectati_actual = 0
        self.nr_vindecati = 0
        self.nr_decese = 0
        self.r0 = params["r0"]
        self.stare_medie = params["stare_medie"]
        self.sever = params["sever"]
        self.rata_mortalitate = params["rata_mortalitate"]
        self.serial_interval = params["serial_interval"]

        self.mediu_rapid = params["incubare"] + params["vindecare"][0]
        self.mediu_lent = params["incubare"] + params["vindecare"][1]
        self.sever_rapid = params["incubare"] + params["vindecare_severa"][0]
        self.sever_lent = params["incubare"] + params["vindecare_severa"][1]
        self.moarte_rapida = params["incubare"] + params["moarte_severa"][0]
        self.moarte_lenta = params["incubare"] + params["moarte_severa"][1]

        self.mediu = {i: {"thetas": [], "rs": []} for i in range(self.mediu_rapid, 365)}
        self.severa = {
            "vindecare": {i: {"thetas": [], "rs": []} for i in range(self.sever_rapid, 365)},
            "deces": {i: {"thetas": [], "rs": []} for i in range(self.moarte_rapida, 365)}
        }

        self.expus_inainte = 0
        self.expus_dupa = 1

        self.populatie_initiala()

    def populatie_initiala(self):
        populatie = 4500
        self.nr_infectati_actual = 1
        self.total_nr_infectati = 1
        indici = np.arange(0, populatie) + 0.5
        self.thetas = np.pi * (1 + 5 ** 0.5) * indici
        self.rs = np.sqrt(indici / populatie)
        self.plot = self.axes.scatter(self.thetas, self.rs, s=5, color=GREY)
        # primul pacient
        self.axes.scatter(self.thetas[0], self.rs[0], s=5, color=RED)
        self.mediu[self.mediu_rapid]["thetas"].append(self.thetas[0])
        self.mediu[self.mediu_rapid]["rs"].append(self.rs[0])

    def raspandire_virus(self, i):
        self.expus_inainte = self.expus_dupa
        if self.day % self.serial_interval == 0 and self.expus_inainte < 4500:
            self.num_new_infected = round(self.r0 * self.total_nr_infectati)
            self.expus_dupa += round(self.num_new_infected * 1.1)
            if self.expus_dupa > 4500:
                self.num_new_infected = round((4500 - self.expus_inainte) * 0.9)
                self.expus_dupa = 4500
            self.nr_infectati_actual += self.num_new_infected
            self.total_nr_infectati += self.num_new_infected
            self.index_noi_infectati = list(
                np.random.choice(
                    range(self.expus_inainte, self.expus_dupa),
                    self.num_new_infected,
                    replace=False))
            thetas = [self.thetas[i] for i in self.index_noi_infectati]
            rs = [self.rs[i] for i in self.index_noi_infectati]
            self.anim.event_source.stop()
            if len(self.index_noi_infectati) > 24:
                size_list = round(len(self.index_noi_infectati) / 24)
                theta_chunks = list(self.chunks(thetas, size_list))
                r_chunks = list(self.chunks(rs, size_list))
                self.anim2 = ani.FuncAnimation(
                    self.fig,
                    self.unul_cate_unul,
                    interval=50,
                    frames=len(theta_chunks),
                    fargs=(theta_chunks, r_chunks, RED))
            else:
                self.anim2 = ani.FuncAnimation(
                    self.fig,
                    self.unul_cate_unul,
                    interval=50,
                    frames=len(thetas),
                    fargs=(thetas, rs, RED))
            self.assign_symptoms()

        self.day += 1

        self.update_status()
        self.update_text()

    def unul_cate_unul(self, i, thetas, rs, color):
        self.axes.scatter(thetas[i], rs[i], s=5, color=color)
        if i == (len(thetas) - 1):
            self.anim2.event_source.stop()
            self.anim.event_source.start()

    def chunks(self, a_list, n):
        for i in range(0, len(a_list), n):
            yield a_list[i:i + n]

    def assign_symptoms(self):
        num_mediu = round(self.stare_medie * self.num_new_infected)
        num_severa = round(self.sever * self.num_new_infected)
        # alege la întâmplare un subset de nou infectați care să aibă simptome medii
        self.mediu_indici = np.random.choice(self.index_noi_infectati, num_mediu, replace=False)
        # atribuie celorlalti simptome severe, rezultatul fiind fie în vindecare, fie în deces
        remaining_indici = [i for i in self.index_noi_infectati if i not in self.mediu_indici]
        sever_vindecare = 1 - (self.rata_mortalitate / self.sever)
        num_vindecare_severa = round(sever_vindecare * num_severa)
        self.severa_indici = []
        self.deces_indici = []
        if remaining_indici:
            self.severa_indici = np.random.choice(remaining_indici, num_vindecare_severa, replace=False)
            self.deces_indici = [i for i in remaining_indici if i not in self.severa_indici]

        # atribuie o zi de vindecare/deces
        low = self.day + self.mediu_rapid
        high = self.day + self.mediu_lent
        for mediu in self.mediu_indici:
            zi_vindecare = np.random.randint(low, high)
            mediu_theta = self.thetas[mediu]
            mediu_r = self.rs[mediu]
            self.mediu[zi_vindecare]["thetas"].append(mediu_theta)
            self.mediu[zi_vindecare]["rs"].append(mediu_r)
        low = self.day + self.sever_rapid
        high = self.day + self.sever_lent
        for vindecare in self.severa_indici:
            zi_vindecare = np.random.randint(low, high)
            vindecare_theta = self.thetas[vindecare]
            vindecare_r = self.rs[vindecare]
            self.severa["vindecare"][zi_vindecare]["thetas"].append(vindecare_theta)
            self.severa["vindecare"][zi_vindecare]["rs"].append(vindecare_r)
        low = self.day + self.moarte_rapida
        high = self.day + self.moarte_lenta
        for deces in self.deces_indici:
            zi_deces = np.random.randint(low, high)
            deces_theta = self.thetas[deces]
            deces_r = self.rs[deces]
            self.severa["deces"][zi_deces]["thetas"].append(deces_theta)
            self.severa["deces"][zi_deces]["rs"].append(deces_r)

    def update_status(self):
        if self.day >= self.mediu_rapid:
            mediu_thetas = self.mediu[self.day]["thetas"]
            mediu_rs = self.mediu[self.day]["rs"]
            self.axes.scatter(mediu_thetas, mediu_rs, s=5, color=GREEN)
            self.nr_vindecati += len(mediu_thetas)
            self.nr_infectati_actual -= len(mediu_thetas)
        if self.day >= self.sever_rapid:
            vindecat_thetas = self.severa["vindecare"][self.day]["thetas"]
            vindecat_rs = self.severa["vindecare"][self.day]["rs"]
            self.axes.scatter(vindecat_thetas, vindecat_rs, s=5, color=GREEN)
            self.nr_vindecati += len(vindecat_thetas)
            self.nr_infectati_actual -= len(vindecat_thetas)
        if self.day >= self.moarte_rapida:
            deces_thetas = self.severa["deces"][self.day]["thetas"]
            deces_rs = self.severa["deces"][self.day]["rs"]
            self.axes.scatter(deces_thetas, deces_rs, s=5, color=BLACK)
            self.nr_decese += len(deces_thetas)
            self.nr_infectati_actual -= len(deces_thetas)

    def update_text(self):
        self.ziua.set_text("Ziua {}".format(self.day))
        self.infectati.set_text("Infectati: {}".format(self.nr_infectati_actual))
        self.decedati.set_text("\nDecedati: {}".format(self.nr_decese))
        self.vindecati.set_text("\n\nVindecati: {}".format(self.nr_vindecati))

    def gen(self):
        while self.nr_decese + self.nr_vindecati < self.total_nr_infectati:
            yield

    def animate(self):
        self.anim = ani.FuncAnimation(
            self.fig,
            self.raspandire_virus,
            frames=self.gen,
            repeat=True)
        self.anim.save('animation.gif', writer='imagemagick', fps=100)



def main():
    coronavirus = Virus(COVID19_PARAMS)
    coronavirus.animate()
    plt.show()
    end = time.time()
    final = end-start
    print (" total: %f s" % final) 
    perIter = (final/102)*1000
    print (" medie per iteratie: %f ms" % perIter) 

if __name__ == "__main__":
    main()
```

```python
from random import randint, random
from matplotlib import pyplot as plt
import time
start = time.time()

class Person:
    def __init__(self, status):
        self.status = status
        self.time_sick = 0
        self.building = list()


class Building:
    def __init__(self, building_id):
        self.building_id = id
        self.num_sick = 0


def simulate(buildings, n_healthy, n_sick, iterations):
    town = [Building(i) for i in range(buildings)]
    people = [Person('healthy') for i in range(n_healthy)] + \
             [Person('sick') for i in range(n_sick)]

    healthy_history = [n_healthy]
    sick_history = [n_sick]
    recovered_history = [0]
    dead_history = [0]
    days = [day for day in range(iterations)]

    for day in days:
        healthy = 0
        sick = 0
        recovered = 0
        dead = 0

        for person in people:
            person.building = town[randint(0, len(town)-1)]

        for person in people:
            if person.status == 'healthy':
                healthy += 1
            elif person.status == 'sick':
                sick += 1
                person.building.num_sick += 1
            elif person.status == 'recovered':
                recovered += 1
            else:
                dead += 1

        for person in people:
            if person.status == 'sick' and person.time_sick < 15:
                person.time_sick += 1
            elif person.status == 'sick' and person.time_sick == 15:
                if randint(0, 9) == 4:
                    person.status = 'dead'
                else:
                    person.status = 'recovered'
            if person.status == 'healthy':
                chance_of_infection = 0.0008*person.building.num_sick
                if random() < chance_of_infection:
                    person.status = 'sick'

        for building in town:
            building.num_sick = 0

        healthy_history.append(healthy)
        sick_history.append(sick)
        recovered_history.append(recovered)
        dead_history.append(dead)

    fig, axs = plt.subplots(4, sharex=True, sharey=False)
    fig.suptitle('SIMULARE')
    axs[0].plot(days, healthy_history[:-1])
    axs[0].set_title('SANATOSI')
    axs[1].plot(days, sick_history[:-1], 'tab:orange')
    axs[1].set_title('BOLNAVI')
    axs[2].plot(days, recovered_history[:-1], 'tab:green')
    axs[2].set_title('RECUPERATI')
    axs[3].plot(days, dead_history[:-1], 'tab:red')
    axs[3].set_title('DECEDATI')

    plt.show()


simulate(75, 4500, 25, 102)
end = time.time()
final = end-start
print (" total: %f s" % final) 
perIter = (final/102)*1000
print (" medie per iteratie: %f ms" % perIter) 
```
